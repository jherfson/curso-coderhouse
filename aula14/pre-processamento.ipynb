## **Deteco e tratamento de Outliers**
Quando os valores **esto muito longe da massa** de dados central, os chamamos de **valores anmalos ou outliers**.

## **Outliers em uma varivel**
Se houver outliers em uma varivel, **podemos v-los no diagrama de caixa e bigodes**.



**_Os Whiskers marcam os valores do inliers mais extremos!_**

### **Como essa separao entre inliers e outliers  estabelecida?**

* Dissemos que uma barreira foi estabelecida, marcando o limite entre inliers e outliers, e no aparecendo no diagrama.
* Para fazer isso, **tomamos o limite da caixa e adicionamos o comprimento de uma caixa e meia. Isso  feito tanto para o limite superior da caixa quanto para o inferior**.

### **Elementos**



_O **comprimentos da caixa**  a diferena entre os quartis 1 e 3 comumente chamados de $Q_{1}(24\%)$ e $Q_{3}(75\%)$, por isso  chamado **amplitude interquartil. Todos os valores fora desses limites so considerados outliers**_



### **Outliers em duas dimenses**
* Pode parecer fcil encontrar outliers em duas dimenses: podemos simplesmente traar diagramas de boxplot para cada uma das duas variveis.
* Isso pode funcionar corretamente na maioria das vezes,  **mas existem distribuies de variveis que desafiam essa anlise**.

## **PCA ou ACP**

* Tcnica   multivariada de modelagem da estrutura de covarincia
* Consiste em transformar um conjunto de variveis originais em outro conjunto de mesma dimenso denominadas de componentes principais (PC ou CP).
* Os   componentes   principais   apresentam propriedades importantes:
    * Cada CP  uma combinao linear de todas as variveis originais, so independentes entre si e estimados com o propsito de reter, em ordem de estimao, o mximo de informao em termos da variabilidade.
    * Essas combinaes   lineares   so   chamadas   de componentes principais e no so correlacionadas entre si.
* Utilizadas   para resolver problemas como reduo da dimensionalidade das variveis, agrupar as observaes pelas similaridades.
* Tcnica reservada para variveis numricas.
* Ajuda em problemas de multicolinearidade.

Sendo cada CP uma combinao linear de variveis originais, o primeiro componente apresenta o maior percentual de varincia explicada.

## **O que fizemos?**

* Simplesmente criamos um novo conjunto de dados.
* Essas novas variveis so calculadas de tal forma que cada uma pode explicar a variabilidade dos dados com diferentes graus de importncia.
* O PC1 explicar a maior porcentagem dos dados, o PC2 um pouco menos e assim por diante em ordem decrescente.
*  uma seleo de variveis muito conveniente para grficos em duas dimenses:

Temos informaes resumidas de um conjunto de dados de 4 variveis originais em 2 componentes principais que mostram claramente a separao dos dados em dois subconjuntos bem definidos.

## **Algumas observao**
* Os componentes principais no tm significado:  um erro tentar interpret-los.
* Duas situaes mais comuns:
    * Se um dado for outlier nos componentes principais, ele tambm ser outlier no conjunto de dados original.
    * Se um conjunto de dados for um grupo separado, ele tambm formar um grupo diferenciado nas variveis originais.

```python
pca = PCA(n_components = 4)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ["PC1", "PC2","PC3","PC4"])
```

```python
principalDf
```

```python
pca.explained_variance_ratio_
```

```python
# Avaliando a variao acumulada dos dois PC
sum(pca.explained_variance_ratio_[0:2])*100
```

```python
sns.set_theme(style="ticks")
sns.pairplot(principalDf)
```

```python
sns.heatmap(principalDf.corr())
```

```python
loadings = pd.DataFrame(pca.components_.T, columns=["PC1", "PC2","PC3","PC4"], index=df.columns[:-1])
loadings
```

```python
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

loading_matrix = pd.DataFrame(loadings, columns=["PC1", "PC2","PC3","PC4"], index=df.columns[:-1])
loading_matrix
```

```python
finalDf = pd.concat([principalDf, df[['target']]], axis = 1)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['target'] == target
    ax.scatter(finalDf.loc[indicesToKeep,"PC1"]
               , finalDf.loc[indicesToKeep, "PC2"]
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

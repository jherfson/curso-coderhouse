## **Detecção e tratamento de Outliers**
Quando os valores **estão “muito” longe da massa** de dados central, os chamamos de **valores anômalos ou outliers**.

## **Outliers em uma variável**
Se houver outliers em uma variável, **podemos vê-los no diagrama de caixa e bigodes**.



**_Os Whiskers marcam os valores do inliers mais extremos!_**

### **Como essa separação entre inliers e outliers é estabelecida?**

* Dissemos que uma barreira foi estabelecida, marcando o limite entre inliers e outliers, e não aparecendo no diagrama.
* Para fazer isso, **tomamos o limite da caixa e adicionamos o comprimento de uma caixa e meia. Isso é feito tanto para o limite superior da caixa quanto para o inferior**.

### **Elementos**



_O **comprimentos da caixa** é a diferença entre os quartis 1 e 3 comumente chamados de $Q_{1}(24\%)$ e $Q_{3}(75\%)$, por isso é chamado **amplitude interquartil. Todos os valores fora desses limites são considerados outliers**_



### **Outliers em duas dimensões**
* Pode parecer fácil encontrar outliers em duas dimensões: podemos simplesmente traçar diagramas de boxplot para cada uma das duas variáveis.
* Isso pode funcionar corretamente na maioria das vezes, 👉 **mas existem distribuições de variáveis que desafiam essa análise**.

Se olharmos apenas os valores da variável x, notamos que o seu diagrama de caixa e bigodes horizontal **tem dois outliers no seu extremo direito**, que correspondem aos dois pontos do gráfico localizados no canto inferior direito. ## **PCA ou ACP**

* Técnica   multivariada de modelagem da estrutura de covariância
* Consiste em transformar um conjunto de variáveis originais em outro conjunto de mesma dimensão denominadas de componentes principais (PC ou CP).
* Os   componentes   principais   apresentam propriedades importantes:
    * Cada CP é uma combinação linear de todas as variáveis originais, são independentes entre si e estimados com o propósito de reter, em ordem de estimação, o máximo de informação em termos da variabilidade.
    * Essas combinações   lineares   são   chamadas   de componentes principais e não são correlacionadas entre si.
* Utilizadas   para resolver problemas como redução da dimensionalidade das variáveis, agrupar as observações pelas similaridades.
* Técnica reservada para variáveis numéricas.
* Ajuda em problemas de multicolinearidade.

* Autovetores - Peso recebido por cada variável

Sendo cada CP uma combinação linear de variáveis originais, o primeiro componente apresenta o maior percentual de variância explicada.

## **O que fizemos?**

* Simplesmente criamos um novo conjunto de dados.
* Essas novas variáveis são calculadas de tal forma que cada uma pode explicar a variabilidade dos dados com diferentes graus de importância.
* O PC1 explicará a maior porcentagem dos dados, o PC2 um pouco menos e assim por diante em ordem decrescente.
* É uma seleção de variáveis muito conveniente para gráficos em duas dimensões:

Temos informações resumidas de um conjunto de dados de 4 variáveis ​​originais em 2 componentes principais que mostram claramente a separação dos dados em dois subconjuntos bem definidos.

## **Algumas observação**
* Os componentes principais não têm significado: é um erro tentar interpretá-los.
* Duas situações mais comuns:
    * Se um dado for outlier nos componentes principais, ele também será outlier no conjunto de dados original.
    * Se um conjunto de dados for um grupo separado, ele também formará um grupo diferenciado nas variáveis originais.

```python
pca = PCA(n_components = 4)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ["PC1", "PC2","PC3","PC4"])
```

```python
principalDf
```

```python
pca.explained_variance_ratio_
```

```python
# Avaliando a variação acumulada dos dois PC
sum(pca.explained_variance_ratio_[0:2])*100
```

```python
sns.set_theme(style="ticks")
sns.pairplot(principalDf)
```

```python
sns.heatmap(principalDf.corr())
```

```python
loadings = pd.DataFrame(pca.components_.T, columns=["PC1", "PC2","PC3","PC4"], index=df.columns[:-1])
loadings
```

```python
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

loading_matrix = pd.DataFrame(loadings, columns=["PC1", "PC2","PC3","PC4"], index=df.columns[:-1])
loading_matrix
```

```python
finalDf = pd.concat([principalDf, df[['target']]], axis = 1)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['target'] == target
    ax.scatter(finalDf.loc[indicesToKeep,"PC1"]
               , finalDf.loc[indicesToKeep, "PC2"]
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

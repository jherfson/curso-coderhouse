## **Data Management Maturity Model**

Na indstria, a administrao de dados passa por 5 nveis que definem sua maturidade.

### **Nvel 1**
* Pouca ou nenhuma governana.
* Funes definidas dentro dos repositrios (data silos).
* Problemas no abordados sobre a qualidade de dados.

### **Nvel 2**
* Gesto inicial.
* Instruo consistente de ferramentas.
* Algumas funes e processos definidos.
* Crescente noo dos impactos dos problemas de qualidade de dados.

### **Nvel 3**
* Data vista como um habilitador organizacional.
* Processos e ferramentas escalveis.
* Metas estabelecidas considerando a qualidade dos dados.
* Automatizao de processos.

### **Nvel 4**
* Gesto centralizada e planejada.
* Gesto de Riscos associados aos dados.
* Mtricas de Performance de Iniciativas de dados.
* Mtricas de melhoria de Qualidade dos dados.

### **Nvel 5**
* Processos com Alta Previso.
* Risco Reduzido.
* Mtricas bem estabelecidas e implementadas para medir a qualidade dos dados.

## **Leitura de arquivos simples com Pandas (Data e Hora)**

````python
from datetime import datetime
````

````python
custom_date_parser = lambda x: datetime.strptime(x, '%Y-%d-%m %H:%M:%S')

df = pd.read_csv(
    'https://raw.githubusercontent.com/jherfson/curso-coderhouse/main/aula11/dates3.txt',
    parse_dates=['date'],
    date_format=custom_date_parser
)
df.info()
````

## **Leitura de arquivos simples com Pandas (Booleanos)**

````python
import pandas as pd
````

1. Ler o arquivos enquete.txt, disponível na pasta da aula. Cuidado, contém espaços em branco.

````python
enquete_df = pd.read_csv(
    'https://raw.githubusercontent.com/jherfson/curso-coderhouse/main/aula11/enquete.txt',
    header=None,
    delim_whitespace=True,
)
enquete_df.head()
````

2. Definir quais variáveis podem ser booleanos (0 ou 1 para True ou False).
3. Converter booleanos diretamente na leitura read_csv.

````python
enquete_df = pd.read_csv(
    'https://raw.githubusercontent.com/jherfson/curso-coderhouse/main/aula11/enquete.txt',
    header=None,
    sep='\s+',
    dtype = {
        'asistio':bool,
        'Tarea':bool
    }
)
enquete_df
````

4. Começar a ler a partir da linha 5 mantendo o header origainal.

````python
enquete_df = pd.read_csv(
    'https://raw.githubusercontent.com/jherfson/curso-coderhouse/main/aula11/enquete.txt',
    header=None,
    sep='\s+',
    dtype = {
        'asistio':bool,
        'Tarea':bool
    },
    skiprows = [i for i in range(1, 5)]
)
enquete_df
````

5. Importar apenas as colunas que não contém NaN

````python
enquete_df = pd.read_csv(
    'https://raw.githubusercontent.com/jherfson/curso-coderhouse/main/aula11/enquete.txt',
    header=None,
    sep='\s+',
    dtype = {
        'asistio':bool,
        'Tarea':bool
    },
    skiprows = [i for i in range(1, 5)],
    usecols=[0, 1, 1, 4]
)
enquete_df
````

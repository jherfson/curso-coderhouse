```python
import os
import pandas as pd
import sqlalchemy as db
from dotenv import load_dotenv
```

```python
load_dotenv()
caminho = os.getenv("DIRETORIO")
nome_arquivo = 'nba_salary.sqlite'
arquivo = os.path.join(caminho, nome_arquivo)

engine = db.create_engine('sqlite:///' + arquivo)
connection = engine.connect()
```

1. Ler o dataset utilizado para extrair apenas os jogadores com um salário maior que 10 milhões de dólares.

```python
sql = "SELECT name FROM sqlite_master WHERE type='table'"

df = pd.read_sql_query(sql, connection)
df
```

```python
sql_salario = '''
SELECT
	*
FROM
	NBA_season1718_salary nss
WHERE
	nss.season17_18 > 10000000;

'''
df = pd.read_sql_query(sql_salario, connection)
df
```

```python
sql_salario = '''
SELECT
	nss.Player,
	nss.season17_18
FROM
	NBA_season1718_salary nss
WHERE
	nss.season17_18 > 10000000;

'''
df = pd.read_sql_query(sql_salario, connection)
df
```

2. Extrair as estatísticas desses jogadores correspondentes ao ano fornecido na tabela Season_Stats do ano 2017.

```python
sql_estatistica_2017 = '''
SELECT
	*
FROM
	Seasons_Stats ss
WHERE
	ss."Year" = 2017
'''
df = pd.read_sql_query(sql_estatistica_2017, connection)
df
```

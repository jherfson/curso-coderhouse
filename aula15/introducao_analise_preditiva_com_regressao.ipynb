# **Introdução à análise preditiva com regressão**

## **Ponto de partida**

🔎 Levantamos a hipótese de que **podia existir algum tipo de dependência** entre as variáveis.

🧐 Se existe esse tipo de dependência, **queremos ver como essa relação ocorre**.

Vamos supor que tenhamos duas variáveis, x e y, agora vejamos o gráfico abaixo:



Nesse caso, parece que as variáveis têm uma forte correlação positiva, e se pensarmos nisso em termos de dependência, significa que quando a variável x aumenta, o mesmo ocorre com a variável y, e vice-versa.

**Atenção:** quando propomos que antes de uma mudança na variável x ocorre uma mudança na variável y. Isso será chamado de **dependência da variável y** em relação à **x**.

## **Como uma função matemática padrão**

* $y = f(x)$ > onde a variável $y$ é uma função de $x$, ou seja, **em última análise, $y$ depende da mudança de $x$**.
* Isso significa que **$x$** é uma variável independente, isto é, **sua mudança não depende do nosso modelo**.

## **Função Linear**

* $y = a + bx$ em que $a$ e $b$ são números reais.
* Esta função produz uma reta no plano.
* O valor de $a$ (ordenada à origem) **mostra que é o valor de $y$ quando valor de $x$ vale $0$**.
* O valor de $B$ (inclinação), por sua vez, **indica o grau de inclinação da reta**.

### **Vejamos dois exemplos**
O **mesmo valor de $a$ com diferentes valores de $b$**, aqui muda a pendente ou inclinação


Um **valor fixo de $b$ para diferentes valores de $a$**, aqui a posição da reta muda, mas sua inclinação permanece a mesma.



## **Ideias prévias**

**Se temos um conjunto de pontos nas variáveis x e y, e de alguma forma y depende de x, uma forma é traçar uma reta que possa representar esses pontos, tomando um critério para a representação e traçando uma reta que o satisfaça**.

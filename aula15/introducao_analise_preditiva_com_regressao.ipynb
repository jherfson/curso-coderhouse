## **Como uma funo matemtica padro**

* $y = f(x)$ > onde a varivel $y$  uma funo de $x$, ou seja, **em ltima anlise, $y$ depende da mudana de $x$**.
* Isso significa que **$x$**  uma varivel independente, isto , **sua mudana no depende do nosso modelo**.

## **Funo Linear**

* $y = a + bx$ em que $a$ e $b$ so nmeros reais.
* Esta funo produz uma reta no plano.
* O valor de $a$ (ordenada  origem) **mostra que  o valor de $y$ quando valor de $x$ vale $0$**.
* O valor de $B$ (inclinao), por sua vez, **indica o grau de inclinao da reta**.

### **Vejamos dois exemplos**
O **mesmo valor de $a$ com diferentes valores de $b$**, aqui muda a pendente ou inclinao


Um **valor fixo de $b$ para diferentes valores de $a$**, aqui a posio da reta muda, mas sua inclinao permanece a mesma.
